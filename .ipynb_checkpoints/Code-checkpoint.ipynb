```python
from PyQt5 import QtWidgets, QtCore
from PyQt5.QtGui import *
from PyQt5.QtWidgets import *
from output2 import Ui_MainWindow
import sys
import cv2
```

```python
def load_image(filenames):
    global started
    global original_img
    started=1
    global pixmap_Original
    original_img = cv2.imread(filenames[0])
    pixmap_Original = QPixmap(filenames[0])
    pixmap_Original = pixmap_Original.scaled(530, 530,aspectRatioMode =1);
    ui.label.setPixmap(pixmap_Original)
    ui.comboBox.setCurrentIndex(0)
    Master_Changer()
    
```

```python
def browse():
    global filenames
    dlg = QFileDialog(filter="Images (*.png , *.jpg , *.jpeg)")
    dlg.setFileMode(QFileDialog.AnyFile)
    if dlg.exec_():
        filenames = dlg.selectedFiles()
        load_image(filenames)
```

```python
def except_hook(cls, exception, traceback):
    sys.__excepthook__(cls, exception, traceback)
```

```python
def Master_Changer():
    global pixmap_Original
    if(started):
        mode=ui.comboBox.currentText()
        #print(mode)
        if mode=="Original":
            ui.label.setPixmap(pixmap_Original)
            ui.verticalSlider.hide()
            ui.verticalSlider_2.hide()
            ui.verticalSlider_3.hide()
            ui.label_2.hide()
            ui.label_3.hide()
            ui.label_4.hide()
        if mode == "Pencil Sketch":
            ui.verticalSlider_3.hide()
            ui.label_4.hide()
            ui.verticalSlider.setMinimum(1)
            ui.verticalSlider.setMaximum(99)
            ui.verticalSlider.setSingleStep(2)
            ui.verticalSlider.setValue(25)
            ui.verticalSlider.show()
            ui.verticalSlider_2.setMinimum(0)
            ui.verticalSlider_2.setMaximum(300)
            ui.verticalSlider_2.setValue(250)
            ui.verticalSlider_2.show()
            ui.label_2.show()
            ui.label_3.show()
            Pencil_Sketch()
        if mode=="Detail Enhancement":
            
            ui.verticalSlider.setMinimum(1)
            ui.verticalSlider.setMaximum(21)
            ui.verticalSlider.setSingleStep(2)
            ui.verticalSlider.setValue(3)
            ui.verticalSlider.show()
        
            ui.verticalSlider_2.setMinimum(3)
            ui.verticalSlider_2.setMaximum(99)
            ui.verticalSlider_2.setValue(5)
            ui.verticalSlider_2.setSingleStep(2)
            ui.verticalSlider_2.show()
            
          
            ui.verticalSlider_3.setMinimum(0)
            ui.verticalSlider_3.setMaximum(100)
            ui.verticalSlider_3.setValue(50)
            ui.verticalSlider_3.show()
            
            ui.label_2.show()
            ui.label_3.show()
            ui.label_4.show()
            
        if mode=="Bilateral Filter":
            ui.verticalSlider.setMinimum(1)
            ui.verticalSlider.setMaximum(21)
            ui.verticalSlider.setSingleStep(2)
            ui.verticalSlider.setValue(3)
            ui.verticalSlider.show()
        
            ui.verticalSlider_2.setMinimum(3)
            ui.verticalSlider_2.setMaximum(99)
            ui.verticalSlider_2.setValue(5)
            ui.verticalSlider_2.setSingleStep(2)
            ui.verticalSlider_2.show()
            
              
            ui.verticalSlider_3.setMinimum(1)
            ui.verticalSlider_3.setMaximum(100)
            ui.verticalSlider_3.setValue(50)
            ui.verticalSlider_3.show()
            
            ui.label_2.show()
            ui.label_3.show()
            ui.label_4.show()
            
            Bilateral_Filter()
            
        if mode== "Pencil Edges":
            
            ui.verticalSlider.setMinimum(1)
            ui.verticalSlider.setMaximum(99)
            ui.verticalSlider.setSingleStep(2)
            ui.verticalSlider.setValue(25)
            ui.verticalSlider.show()
        
            ui.verticalSlider_2.setMinimum(3)
            ui.verticalSlider_2.setMaximum(9)
            ui.verticalSlider_2.setValue(3)
            ui.verticalSlider_2.setSingleStep(2)
            ui.verticalSlider_2.show()
            
              
            ui.verticalSlider_3.setMinimum(10)
            ui.verticalSlider_3.setMaximum(255)
            ui.verticalSlider_3.setValue(150)
            ui.verticalSlider_3.show()
            
            ui.label_2.show()
            ui.label_3.show()
            ui.label_4.show()
            
            
            
            Pencil_Edges()

            
def Master_Changer_2():
    global pixmap_Original
    if(started):
        mode=ui.comboBox.currentText()
        #print(mode)
        if mode == "Pencil Sketch":
            Pencil_Sketch()
        if mode=="Detail Enhancement":
            Detail_Enhancement()
        if mode=="Bilateral Filter":
            Bilateral_Filter()
        if mode== "Pencil Edges":
            Pencil_Edges()

#Bilateral Filter
#Pencil Edges
```

```python
def Pencil_Sketch():
    global original_img
    kernel=ui.verticalSlider.value()
    scale=ui.verticalSlider_2.value()
    if(kernel%2==0):
        kernel+=1
        
    ui.label_2.setText("Kernel\n"+str(kernel))
    ui.label_3.setText("Scale\n"+str(scale))
    # Convert the image into grayscale image
    gray = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
   
        # Blur the image using Gaussian Blur 
    gray_blur = cv2.GaussianBlur(gray, (kernel, kernel), 0)

       # Convert the image into pencil sketch
    cartoon = cv2.divide(gray, gray_blur, scale=scale)
    cv2.imwrite("Pencil_Sketch.png", cartoon)
    pixmap = QPixmap('Pencil_Sketch.png')
    pixmap = pixmap.scaled(530, 530,aspectRatioMode =1);
    ui.label.setPixmap(pixmap)
```

```python
def Detail_Enhancement():
    global original_img
    kernel=ui.verticalSlider.value()
    smooth=ui.verticalSlider_2.value()
    edge_p=ui.verticalSlider_3.value()
    edge_p/=100
    print(edge_p)
    if(kernel%2==0):
        kernel+=1
    if(smooth%2==0):
        smooth+=1
    
    ui.label_2.setText("Kernel\n"+str(kernel))
    ui.label_3.setText("Smooth\n"+str(smooth))
    ui.label_4.setText("Edge\n"+str(edge_p))
    gray = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
    gray = cv2.medianBlur(gray, kernel) 
    edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9) 
    color = cv2.detailEnhance(original_img, sigma_s=smooth, sigma_r=edge_p)
    cartoon = cv2.bitwise_and(color, color, mask=edges)    
    cv2.imwrite("Detail_Enhancement.png", cartoon)
    pixmap = QPixmap('Detail_Enhancement.png')
    pixmap = pixmap.scaled(530, 530,aspectRatioMode =1);
    ui.label.setPixmap(pixmap)
```

```python
def Bilateral_Filter():
    global original_img
    kernel=ui.verticalSlider.value()
    smooth=ui.verticalSlider_2.value()
    edge_preserve =ui.verticalSlider_3.value()
    if(kernel%2==0):
        kernel+=1
    if(smooth%2==0):
        smooth+=1
    
    ui.label_2.setText("Kernel\n"+str(kernel))
    ui.label_3.setText("Smooth\n"+str(smooth))
    ui.label_4.setText("Edge\n"+str(edge_preserve))
    
    gray = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
    gray = cv2.medianBlur(gray, kernel) 
    edges = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
    
    color = cv2.bilateralFilter(original_img, smooth, edge_preserve, smooth) 
    cartoon = cv2.bitwise_and(color, color, mask=edges) 
    
    cv2.imwrite("Bilateral_Filter.png", cartoon)
    pixmap = QPixmap('Bilateral_Filter.png')
    pixmap = pixmap.scaled(530, 530,aspectRatioMode =1);
    ui.label.setPixmap(pixmap)
```

```python
def Pencil_Edges():
    global original_img
    kernel=ui.verticalSlider.value()
    laplacian_filter =ui.verticalSlider_2.value()
    noise_reduction  =ui.verticalSlider_3.value()
    if(kernel%2==0):
        kernel+=1
    if(laplacian_filter%2==0):
        laplacian_filter+=1
    
    ui.label_2.setText("Kernel\n"+str(kernel))
    ui.label_3.setText("Laplacian\n"+str(laplacian_filter))
    ui.label_4.setText("Noise\n"+str(noise_reduction))
    
    gray = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
    gray = cv2.medianBlur(gray, kernel) 
    edges = cv2.Laplacian(gray, -1, ksize=laplacian_filter)

        
    edges_inv = 255-edges
    
    dummy, cartoon = cv2.threshold(edges_inv, noise_reduction, 255, cv2.THRESH_BINARY)
    
    cv2.imwrite("Pencil_Edges.png", cartoon)
    pixmap = QPixmap('Pencil_Edges.png')
    pixmap = pixmap.scaled(530, 530,aspectRatioMode =1);
    ui.label.setPixmap(pixmap)
```

```python
started=0
app = QApplication(sys.argv)
style = open('Darkeum.qss').read()
app.setStyleSheet(style)
MainWindow = QMainWindow()
ui = Ui_MainWindow()
ui.setupUi(MainWindow)
MainWindow.show()
ui.verticalSlider.valueChanged.connect(Master_Changer_2)
ui.verticalSlider_2.valueChanged.connect(Master_Changer_2)
ui.verticalSlider_3.valueChanged.connect(Master_Changer_2)
#ui.verticalSlider.sliderMoved.connect(Master_Changer)
ui.verticalSlider.hide()
ui.label_2.hide()
ui.verticalSlider_2.hide()
ui.label_3.hide()
ui.verticalSlider_3.hide()
ui.label_4.hide()
ui.comboBox.currentIndexChanged.connect(Master_Changer)
ui.loadButton.clicked.connect(browse)
def except_hook(cls, exception, traceback):
    sys.__excepthook__(cls, exception, traceback)

sys.excepthook = except_hook
sys.exit(app.exec_())

```
